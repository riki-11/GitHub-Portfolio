```python
import pandas as pd
```

## Data Cleaning - clinics ##

```python
df_clinics = pd.read_csv('clinics.csv', encoding="Latin1").drop_duplicates(subset='clinicid').drop_duplicates()
df_clinics.head(10)

# total rows:53962
```

```python
#checking for duplicate keys

containsDuplicate = df_clinics['clinicid'].duplicated().any()
print(containsDuplicate)
```

```python
df_clinics['hospitalname'].fillna("noName", inplace=True)
```

## Data Cleaning - Doctors ##

```python
df_doctors = pd.read_csv('doctors.csv', encoding="Latin1").drop_duplicates(subset='doctorid').drop_duplicates()
df_doctors.head(10)

# total rows:60024
```

```python
# check for NaNs
print(df_doctors.isna().sum())
```

```python
# Checking for Duplicate keys
containsDuplicate = df_doctors['doctorid'].duplicated().any()
print(containsDuplicate)
```

```python
# checking for multiple representations of "mainspecialty" column

pd.set_option('display.max_rows', None)

df_doctors['mainspecialty'].value_counts() 
```

```python
# Handling Nans
df_doctors['age'].fillna(-1, inplace=True)
df_doctors['mainspecialty'].fillna("None", inplace=True)
```

## Data Cleaning - PXID ##

```python
df_px = pd.read_csv('px.csv', encoding="Latin1").drop_duplicates()
df_px.head(10)

# total rows:5,773,951
```

```python
# check for NaNs
print(df_px.isna().sum())
```

```python
# Handling Nans
df_px['age'].fillna(-1, inplace=True)
```

```python
#Checking for Duplicates
print(df_px['pxid'].duplicated().any())
```

```python
#Dropping Duplicates
df_px.drop_duplicates(subset='pxid', keep='first', inplace=True)
```

```python
#Handling Multiple Data Types for Age Column
df_px = df_px[df_px['age'] != 'age']
df_px['age'] = df_px['age'].astype(int)
```

## Data Cleaning - Appointments

```python
df_appointments = pd.read_csv('appointments.csv').dropna()
df_appointments.head(10)

# total rows: 3,293,829 
```

```python
# checking for multiple representations of similar data - Appointments

print(df_appointments.status.unique())

print(df_appointments.type.unique())

print(df_appointments.Virtual.unique())
```

## Dropping Rows from Fact Table where Keys have no match

```python
df_appointments = df_appointments[df_appointments['pxid'].isin(df_px['pxid'])]
df_appointments = df_appointments[df_appointments['clinicid'].isin(df_clinics['clinicid'])]
df_appointments = df_appointments[df_appointments['doctorid'].isin(df_doctors['doctorid'])]
```

```python
# Exporting a Copy of the Cleaned Data:

df_appointments.to_csv('cleaned_data/appointments.csv', index=False)
df_px.to_csv('cleaned_data/px.csv', index=False)
df_doctors.to_csv('cleaned_data/doctors.csv', index=False)
df_clinics.to_csv('cleaned_data/clinics.csv', index=False)
```

```python
# Export every table to csv in a separate folder
df_appointments.to_csv('cleaned_data/appointments.csv', index=False)
df_doctors.to_csv('cleaned_data/doctors.csv', index=False)
df_clinics.to_csv('cleaned_data/clinics.csv', index=False)
df_px.to_csv('cleaned_data/px.csv', index=False)
```

## Loading Transformed Data into Data Warehouse

```python
# Importing Module
import mysql.connector
```

```python
# Connecting to Local MySQL Database
conn = mysql.connector.connect(
    host = "127.0.0.1",
    port = 3306,
    user = "root",
    password = "Pilipinas-19",
    database = 'mco1_dw'
)

cursor = conn.cursor()
```

```python
#Loading Clinics Table

cursor.execute("CREATE TABLE clinics (clinic_id VARCHAR(100), hospital_name VARCHAR(100), is_hospital BOOLEAN, city VARCHAR(100), province VARCHAR(100), region_name VARCHAR(100))")

for row in df_clinics.itertuples():
    cursor.execute("INSERT INTO clinics (clinic_id, hospital_name, is_hospital, city, province, region_name) VALUES (%s,%s, %s, %s, %s, %s)", (row.clinicid, row.hospitalname, row.IsHospital, row.City, row.Province, row.RegionName))

conn.commit()
```

```python
#Loading Doctors Table

cursor.execute("CREATE TABLE doctors (doctor_id VARCHAR(100), main_specialty VARCHAR(500), age INT)")

for row in df_doctors.itertuples():
    cursor.execute("INSERT INTO doctors (doctor_id, main_specialty, age) VALUES (%s,%s, %s)", (row.doctorid, row.mainspecialty, row.age))

conn.commit()
```

```python
#Loading Patients Table

cursor.execute("CREATE TABLE px (px_id VARCHAR(100), age INT, gender VARCHAR(6))")

for row in df_px.itertuples():
    cursor.execute("INSERT INTO px (px_id, age, gender) VALUES (%s,%s, %s)", (row.pxid, row.age, row.gender))

conn.commit()
```

```python
# Loading Appointments Table

cursor.execute("""
    CREATE TABLE appointments (
        px_id VARCHAR(100),
        clinic_id VARCHAR(100),
        doctor_id VARCHAR(100),
        app_id VARCHAR(100),
        status VARCHAR(100),
        time_queued DATETIME,
        queue_date DATETIME,
        start_time DATETIME,
        end_time DATETIME,
        type VARCHAR(100),
        `virtual` BOOLEAN
    )
""")

for row in df_appointments.itertuples():
    cursor.execute("INSERT INTO appointments (px_id, clinic_id, doctor_id, app_id, status, time_queued, queue_date, start_time, end_time, type, `virtual`) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)", (row.pxid, row.clinicid, row.doctorid, row.apptid, row.status, row.TimeQueued, row.QueueDate, row.StartTime, row.EndTime, row.type, row.Virtual))

conn.commit()
```
